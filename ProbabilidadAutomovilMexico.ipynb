## Variable endógena/dependiente
- *(Dummy)* *__El hogar cuenta con al menos un automóvil.__* 
Se modela a partir del número de autos con los que cuenta un hogar.
    - 1 = El hogar cuenta con al menos un auto
    - 0 = El hogar **NO** cuenta con auto

## Variables exógenas/independientes
1. *(Cuantitativa)* *__Ingresos acumulados del hogar__* Se espera que los ingresos acumulados trimestrales sean un indicador de riqueza y proxy de recursos suficientes para tener un vehículo. **Se contempla una relación positiva entre los ingresos acumulados y la probabilidad de contar con al menos un vehículo.**
2. *(Cuantitativa)* *__Número de integrantes del hogar__* . Económicamente un número mayor de integrantes puede significar menos recursos disponibles para tener un vehículo. Socialmente, el mayor número de integrantes puede significar una mayor necesidad para la seguridad y el transporte de la familia. **Bajo una visión económica se espera una relación inversa entre esta variable y la endógena.**
2. *(Cuantitativa)* *__Numero de fuentes de ingresos del hogar__* Un número mayor de fuentes de ingresos pueden indicar, entre tantas cosas, mayores ingresos percibidos, diversificación de recursos e inversiones, múltiples negocios, etc. **Se espera una relación positiva**
3. *(Dummy)* *__El jefe del hogar es hombre__* El sexo puede ser un determinante para la adquisición de un vehículo, ya sea por una desigualdad salarial, el interés de los sexos por contar con un auto o factores como la cultura o la aversión al conducir. **Se contempla una mayor probabilidad esperada al ser hombre por las distintas desigualdades que se perciben en el país.**
    - 1 = Es hombre
    - 0 = Es mujer
4. *(Cuantitativa)* *__Edad del jef(a) del hogar:__* La edad puede ser un buen determinante pues al ser mayor y tener más experiencia, se esperarían mayores oportunidades como lo son las laborales y salariales. Asimismo, la edad puede significar una acumulación de riqueza durante los años en los que se ha laborado. Estos puntos pueden ofrecer los recursos para contar con un automóvil. **Se espera una relación positiva.** 
5. *(Dummy)* *__El jefe del hogar terminó la secundaria__* El haber terminado la secundaria puede ser un indicador de las oportunidades económicas como lo son las laborales y salariales. La retribución por el trabajo de una persona sin la secundaria terminada puede no ser suficientes para contar con un vehículo. **Se contempla un aumento en la probabilidad de tener un auto al terminar la secundaria.**
    - 1 = El/La jef(a) del hogar terminó la secundaria
    - 0 = El/La jef(a) del hogar **NO** terminó la secundaria
8. *(Cuantitativa)* *__Tasa de ahorro__*. Económicamente, una tasa de ahorro mayor implica un menor consumo, por lo que se podría deducir que no se están consumiendo gran cantidad de productos. Dentro de estos podrían entrar los insumos de un vehículo: gasolina, pago de tenencia, pago de multas, reparaciones, etc. Por lo tanto, **se espera una relación inversa entre la tasa de ahorro y contar con un vehículo.**

## Creación de la base de datos del trabajo
Contempla el procesamiento de los datos y el uso de funciones como pivot tables y combinaciones entre bases de datos

```python
# Hogares por variables agregadas (promedio de los registros)
# al ser los mismos valores en distintos registros de un mismo hogar, el promedio de estos da ese mismo valor
hogares_va = pd.pivot_table(hogares, values=['num_auto', 'est_alim', 'est_trans'], index=['folioviv', 'foliohog'],
              aggfunc='mean')

# Hogares Variables argregadas manejadas
hogares_vam = hogares_va.copy()
# Si num_auto es mayor o igual a 1, tiene valor de 1 mientras que si no, tiene valor 0
hogares_vam['num_auto'] = hogares_vam['num_auto'].apply(lambda num_auto: 1 if num_auto >=1 else 0)
# Poner est_alim y est_trans a trimestres, es decir, multiplicarlos por 3 porque estan en meses
hogares_vam[['est_alim','est_trans']] = hogares_vam[['est_alim', 'est_trans']].apply(lambda col: col*3)
# Se cambia el nombre de la columna de est_alim y est_trans a ..._tri de trimestral y de num_auto a tiene_auto
hogares_vam.rename(columns={'est_alim':'est_alim_tri', 'est_trans':'est_trans_tri', 'num_auto':'tiene_auto'}, inplace=True)
hogares_vam
```

```python
# Gasto de los hogares por variables sumadas (suma de todos los gastos por hogar)
hogares_gastos_va = pd.pivot_table(hogares_gastos_m, values='gasto_tri', index=['folioviv', 'foliohog'], aggfunc=np.sum)
# cambio de gasto_tri a gastot_tri para indicar el total de gasto acumulado
hogares_gastos_va.columns = ['gastot_tri']
hogares_gastos_va
```

```python
# Ingresos variables agregadas (conteo de tipos de ingreso y suma de los ingresos)
ingresos_va = pd.pivot_table(ingresos_m, values='ing_tri', index=['folioviv', 'foliohog'],
               aggfunc=[np.sum, 'count'])

# Ingresos variables agregadas manejadas
ingresos_vam = ingresos_va.copy()
# Se cambian las columnas por ingtot_tri: ingreso total trimestral del hogar
# y a num_ingresos: numero de ingresos percibidos en el hogar
ingresos_vam.columns = ['ingtot_tri', 'num_ingresos']
ingresos_vam
```

```python
# Poblacion jefes
poblacion_m
poblacion_jefes = poblacion_m[poblacion_m['parentesco']==101]
poblacion_jefes = poblacion_jefes.astype({'nivelaprob':'int64'})
# Booleano si termino la secundaria, 1 si el nivel es mayor o igual a 3 (secundaria) y 0 si no lo es
poblacion_jefes['termino_sec'] = poblacion_jefes['nivelaprob'].apply(lambda nivel: 1 if nivel >= 3 else 0 )
# Boolean si es hombre, 1 es hombre, 0 es mujer
poblacion_jefes['es_hombre'] = poblacion_jefes['sexo'].apply(lambda sexo: 1 if sexo == 1 else 0)
# Variables de interes de poblacion jefes del hogar
poblacion_jefes_vi = ['folioviv', 'foliohog', 'es_hombre', 'edad', 'termino_sec']
# Poblacion jefes manejada
poblacion_jefes_m = poblacion_jefes[poblacion_jefes_vi].set_index(['folioviv', 'foliohog'])
poblacion_jefes_m.rename(columns={'edad':'edad_jefe'}, inplace=True)
poblacion_jefes_m
```

```python
# Poblacion integrantes
# Es la poblacion por el numero de hombres y mujeres contadas
poblacion_integrantes = pd.pivot_table(poblacion_m, values='numren', index=['folioviv', 'foliohog'], columns='sexo', aggfunc='count')
poblacion_integrantes.columns = ['num_hombres', 'num_mujeres']
# Se reemplazan los NaN con 0 para hacer la suma
poblacion_integrantes.replace(np.nan, 0, inplace=True)
# Se crea la columna num_integrantes, es la suma de hombres mas mujeres
poblacion_integrantes['num_integrantes'] = poblacion_integrantes['num_hombres'] + poblacion_integrantes['num_mujeres']
poblacion_integrantes
```

## Base de datos Final

## Base conjunta propia

```python
# Hogares_trabajo
# Une las bases de datos de hogares_vam y hogares_gastos con base en el index (hogar)
hogares_t = pd.merge(hogares_vam, hogares_gastos_va, left_index=True, right_index=True)
# Un la base de datos de hogares_t con los registros de ingresos_vam con base en el hogar único
hogares_t = hogares_t.merge(ingresos_vam, left_index=True, right_index=True)
# Une la base de datos hogares_t con los registros de poblacion_jefes (datos de los jefes del hogar) 
hogares_t = hogares_t.merge(poblacion_jefes_m, left_index=True, right_index=True)
# Une la base de datos hogares_t con los registros de poblacion_integrantes (datos de los integrantes por hogar único)
hogares_t = hogares_t.merge(poblacion_integrantes, left_index=True, right_index=True)
# Crea una tasa de hombres que viven en lugar con respecto al numero de integrantes
hogares_t['tasa_hombres'] = round(hogares_t['num_hombres'] / hogares_t['num_integrantes'], 2)
# Genera la tasa de ahorro de los hogares
hogares_t['tasa_ahorro_tri'] = round((hogares_t['ingtot_tri']-hogares_t['gastot_tri'])/hogares_t['ingtot_tri'], 3)
# Genera la tasa/proporción que se gasta en alimentos estimados con respecto al ingreso, redondeado a 3 decimas
hogares_t['tasa_alim_tri'] = round(hogares_t['est_alim_tri']/hogares_t['ingtot_tri'], 3)
hogares_t.to_csv('bdd/creados/bdd_hogares.csv')
hogares_t

# Comprobacion suma
tiene = hogares_t[hogares_t['tiene_auto']==1].shape[0]
no_tiene = hogares_t[hogares_t['tiene_auto']==0].shape[0]
concuerda = (hogares_t.shape[0] == tiene + no_tiene)

print('Hogares con auto: {:,} | Hogares SIN auto: {:,} | Suma concuerda: {}'.format(tiene, no_tiene, concuerda))
```
